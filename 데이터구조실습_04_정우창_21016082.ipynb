# 문제1

재귀적 방법을 이용하여 word 가 회문(palindrome)인지를 판별하는 함수를 작성하시오.

str 이 회문이면 True를 리턴하고 회문이 아니면 False를 리턴하는 함수를 작성하시오.


***결과***
```
True
True
False
True
False
```

````python
def is_palindrome(word: str) -> bool:
    if len(word) <= 1:
        return True
    return (word[0] == word[len(word)-1]) and is_palindrome(word[1:len(word)-1])
    

# 아래는 수정하지 마시오.

print(is_palindrome(''))
print(is_palindrome('a'))
print(is_palindrome('abc'))
print(is_palindrome('aabcbaa'))
print(is_palindrome('aabcbba'))
print(is_palindrome('avbaabva'))
            
````

# 문제2

LinkedList 안에 count() 메소드를 구현하시오.

count() 메소드는 매개변수 data와 같은 값을 가지고 있는 노드의 개수를 리턴한다.

***결과***
```
2
0
3
1
```

````python
from __future__ import annotations
from typing import Any

class Node:
    """연결 리스트용 노드 클래스"""
    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedList:
    """연결 리스트 클래스"""
    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.head = None     # 머리 노드

    def __len__(self) -> int:
        """연결 리스트의 노드 개수를 반환"""
        return self.no

    def add_first(self, data: Any) -> None:
        """맨 앞에 노드를 삽입"""
        ptr = self.head  # 삽입 전의 머리 노드
        self.head = Node(data, ptr)
        self.no += 1

    def add_last(self, data: Any):
        """맨 끝에 노드를 삽입"""
        if self.head is None :    # 리스트가 비어 있으면
            self.add_first(data)  # 맨앞에 노드 삽입
        else:
            ptr = self.head
            while ptr.next is not None:
                ptr = ptr.next  # while문을 종료할 때 ptr은 꼬리 노드를 참조
            ptr.next = Node(data, None)
            self.no += 1

    def print(self) -> None:
        """모든 노드를 출력"""
        ptr = self.head

        while ptr is not None:
            print(ptr.data, end='')
            if ptr.next is not None:
                print('->', end='')
            else:
                print()
            ptr = ptr.next
    
    def search(self, data: Any) -> int:
        """data와 값이 같은 노드를 검색"""
        idx = 0
        ptr = self.head
        while ptr is not None:
            if ptr.data == data:
                return idx
            idx += 1
            ptr = ptr.next
        return -1
    
    def __contains__(self, data: Any) -> bool:
        """연결 리스트에 data가 포함되어 있는가?"""
        return self.search(data) >= 0
    
    def count(self, data: Any) -> int:
        """연결 리스트에 data가 포함되어 있는 노드의 개수 리턴"""
        cnt = 0
        ptr = self.head
        while ptr is not None:
            if ptr.data == data:
                cnt += 1
            ptr = ptr.next
        return cnt

        
## 아래는 수정하지 마시오.
my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('A')
my_list.add_last('C')
my_list.add_last('D')
my_list.add_last('C')
my_list.add_last('C')

print(my_list.count('A'))
print(my_list.count('B'))
print(my_list.count('C'))
print(my_list.count('D'))
````

# 문제3

LinkedList 클래스의 remove_last(), remove() 메소드를 구현하시오.

***결과***
```
A->B->C->D
A->B->C
A->C
```

````python
from __future__ import annotations
from typing import Any

class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedList:
    """연결 리스트 클래스"""

    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.head = None     # 머리 노드

    def __len__(self) -> int:
        """연결 리스트의 노드 개수를 반환"""
        return self.no

    def add_first(self, data: Any) -> None:
        """맨 앞에 노드를 삽입"""
        ptr = self.head  # 삽입 전의 머리 노드
        self.head = Node(data, ptr)
        self.no += 1

    def add_last(self, data: Any):
        """맨 끝에 노드를 삽입"""
        if self.head is None :    # 리스트가 비어 있으면
            self.add_first(data)  # 맨앞에 노드 삽입
        else:
            ptr = self.head
            while ptr.next is not None:
                ptr = ptr.next  # while문을 종료할 때 ptr은 꼬리 노드를 참조
            ptr.next = Node(data, None)
            self.no += 1

    def print(self) -> None:
        """모든 노드를 출력"""
        ptr = self.head

        while ptr is not None:
            print(ptr.data, end='')
            if ptr.next is not None:
                print('->', end='')
            else:
                print()
            ptr = ptr.next
    
    def search(self, data: Any) -> int:
        """data와 값이 같은 노드를 검색"""
        idx = 0
        ptr = self.head
        while ptr is not None:
            if ptr.data == data:
                return idx
            idx += 1
            ptr = ptr.next
        return -1
    
    def __contains__(self, data: Any) -> bool:
        """연결 리스트에 data가 포함되어 있는가?"""
        return self.search(data) >= 0

    def remove_first(self) -> None:
        """머리 노드를 삭제"""
        if self.head is not None:  # 리스트가 비어 있지 않으면
            self.head = self.head.next
            self.no -= 1

    def remove_last(self):
        """꼬리 노드 삭제"""
        ptr = self.head
        pre = None
        while ptr and ptr.next is not None:
            pre = ptr
            ptr = ptr.next
        if pre is not None:
            pre.next = None

    def remove(self, p: Node) -> None:
        """노드 p를 삭제"""        
        ptr = self.head
        pre = None
        while ptr is not None:
            if ptr is p:
                break
            pre = ptr
            ptr = ptr.next
            
        if ptr is not None:    
            pre.next = ptr.next
        

    def clear(self) -> None:
        """전체 노드를 삭제"""
        while self.head is not None:  # 전체가 비어 있게 될 때까지
            self.remove_first()       # 머리 노드를 삭제
        self.no = 0

        
## 아래는 수정하지 마시오.
my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('B')
my_list.add_last('C')
my_list.add_last('D')

my_list.print()
my_list.remove_last() # 마지막 노드 제거
my_list.print()
my_list.remove(my_list.head.next) # 머리노드 다음 노드 제거
my_list.print()
````

# 문제4

LinkedList 클래스의 remove_at() 메소드를 완성하시오.

***결과***
```
A->B->C->D
A->B->D
B->D
B
1
```

````python
from __future__ import annotations
from typing import Any, List, Tuple, Dict

class Node:
    """연결 리스트용 노드 클래스"""

    def __init__(self, data: Any = None, next: Node = None):
        """초기화"""
        self.data = data  # 데이터
        self.next = next  # 뒤쪽 포인터


class LinkedList:
    """연결 리스트 클래스"""

    def __init__(self) -> None:
        """초기화"""
        self.no = 0          # 노드의 개수
        self.head = None     # 머리 노드

    def __len__(self) -> int:
        """연결 리스트의 노드 개수를 반환"""
        return self.no

    def add_first(self, data: Any) -> None:
        """맨 앞에 노드를 삽입"""
        ptr = self.head  # 삽입 전의 머리 노드
        self.head = Node(data, ptr)
        self.no += 1

    def add_last(self, data: Any):
        """맨 끝에 노드를 삽입"""
        if self.head is None :    # 리스트가 비어 있으면
            self.add_first(data)  # 맨앞에 노드 삽입
        else:
            ptr = self.head
            while ptr.next is not None:
                ptr = ptr.next  # while문을 종료할 때 ptr은 꼬리 노드를 참조
            ptr.next = Node(data, None)
            self.no += 1

    def print(self) -> None:
        """모든 노드를 출력"""
        ptr = self.head

        while ptr is not None:
            print(ptr.data, end='')
            if ptr.next is not None:
                print('->', end='')
            else:
                print()
            ptr = ptr.next
    
    def search(self, data: Any) -> int:
        """data와 값이 같은 노드를 검색"""
        idx = 0
        ptr = self.head
        while ptr is not None:
            if ptr.data == data:
                return idx
            idx += 1
            ptr = ptr.next
        return -1
    
    def __contains__(self, data: Any) -> bool:
        """연결 리스트에 data가 포함되어 있는가?"""
        return self.search(data) >= 0

    def remove_first(self) -> None:
        """머리 노드를 삭제"""
        if self.head is not None:  # 리스트가 비어 있지 않으면
            self.head = self.head.next
            self.no -= 1

    def remove_at(self, index: int) -> None:
        """특정 위치에 있는 노드를 삭제"""
        if index < 0 or self.no <= index: # out of index
            raise IndexError()
        if index == 0:
            self.head = self.head.next
            self.no -= 1
            return
        ptr = self.head
        pre = None
        cnt = 0
        while ptr is not None:
            if cnt == index:
                break
            pre = ptr
            ptr = ptr.next
            cnt += 1
            
        pre.next = ptr.next
        self.no -= 1
                
    def clear(self) -> None:
        """전체 노드를 삭제"""
        while self.head is not None:  # 전체가 비어 있게 될 때까지
            self.remove_first()       # 머리 노드를 삭제
        self.no = 0

        
## 아래는 수정하지 마시오.
my_list = LinkedList()
my_list.add_last('A')
my_list.add_last('B')
my_list.add_last('C')
my_list.add_last('D')

my_list.print()
my_list.remove_at(2)
my_list.print()
my_list.remove_at(0)
my_list.print()
my_list.remove_at(1)
my_list.print()
print(len(my_list))
````
