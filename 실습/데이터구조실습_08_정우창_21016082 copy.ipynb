## 문제9

배열을 이용한 완전이진트리에서 부모 노드의 키를 리턴하는 함수를 작성하시오.

- 파라미터:
    - cbt: 리스트
    - index: 인덱스
- 리턴:
    - index에 위치한 노드의 부모노드의 키 리턴
    
### 결과

```
None
A
B
E
```

````python
def get_parent(cbt, index):
    if 0<=index//2<len(cbt):
        return cbt[index//2]

cbt = [None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
print(get_parent(cbt, index=1))
print(get_parent(cbt, index=2))
print(get_parent(cbt, index=5))
print(get_parent(cbt, index=11))

````

## 문제10

배열을 이용한 완전이진트리에서 왼쪽자식 노드의 키를 리턴하는 함수를 작성하시오.

- 파라미터:
    - cbt: 리스트
    - index: 인덱스
- 리턴:
    - index에 위치한 노드의 왼쪽 자식 노드의 키 리턴
    
***결과***
```
B
D
J
None
None
```

````python
def get_parent_left_child(cbt, index):
    if 0<=2*index<len(cbt):
        return cbt[2*index]

cbt = [None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
print(get_parent_left_child(cbt, index=1)) # A의 왼쪽
print(get_parent_left_child(cbt, index=2)) # B의 왼쪽
print(get_parent_left_child(cbt, index=5)) # E의 왼쪽
print(get_parent_left_child(cbt, index=6)) # F의 왼쪽
print(get_parent_left_child(cbt, index=11)) # K의 왼쪽
````

## 문제11

배열을 이용한 완전이진트리에서 오른쪽 자식노드의 키를 리턴하는 함수를 작성하시오.

- 파라미터:
    - cbt: 리스트
    - index: 인덱스
- 리턴:
    - index에 위치한 노드의 오른쪽 자식노드의 키

***결과***
```
C
E
K
None
None
```

````python
def get_parent_right_child(cbt, index):
    if 0<=2*index + 1 < len(cbt):
        return cbt[2*index + 1]

cbt = [None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
print(get_parent_right_child(cbt, index=1)) # A의 오른쪽
print(get_parent_right_child(cbt, index=2)) # B의 오른쪽
print(get_parent_right_child(cbt, index=5)) # E의 오른쪽
print(get_parent_right_child(cbt, index=6)) # F의 오른쪽
print(get_parent_right_child(cbt, index=11)) # K의 오른쪽
````

## 문제12

배열을 이용한 완전이진트리에서 노드의 level을 리턴하는 함수를 작성하시오.

- 파라미터:
    - cbt: 리스트
    - index: 인덱스
- 리턴:
    - index에 위치한 노드의 레벨

### 결과

```
0
1
2
2
3
```

````python
def get_level(cbt, index):
    level=-1
    while cbt[index]:
        index //= 2
        level += 1
    return level

cbt = [None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
print(get_level(cbt, index=1)) # A의 level
print(get_level(cbt, index=2)) # B의 level
print(get_level(cbt, index=5)) # E의 level
print(get_level(cbt, index=6)) # F의 level
print(get_level(cbt, index=11)) # K의 level
````
